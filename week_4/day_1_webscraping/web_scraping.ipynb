```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from bs4 import BeautifulSoup
import requests
import re
import json
```

```python
# scrapping movie data from 'the numbers'
def movie_data(num):
    url=requests.get('https://www.the-numbers.com/movie/budgets/all/{}'.format(num))
    return BeautifulSoup(url.content, 'html.parser')  

```

```python
# create emply list for 1000 data
list1=[]
for i in list(range(1,1000,100)):
    list1.append(movie_data(i))
list1
```

```python
# find all data with <tr>
list1[0].find_all('tr')
```

```python
# After that, find all data with <td> and set an empty list to fill in the selected data. 
data=[]
for m in list1:
    k=m.findAll('td')
    for j in k:
        data.append(j.text)
data
```

```python
# each movie contains 6 lines of data, so the total len(data) for 1000 movie is 6000.
# set index to the first item in the list. (len(index) becomes 6000/6=1000)
index=[]
for i in list(range(0,6000,6)):
    index.append(data[i])
len(index)    
```

```python
# set date to the second item in the list.
date=[]
for i in list(range(1,6000,6)):
    date.append(data[i])
date
```

```python
# set name to the third item in the list.
name=[]
for i in list(range(2,6000,6)):
    name.append(data[i])
name
```

```python
# set budget to the fourth item in the list.
budget=[]
for i in list(range(3,6000,6)):
    budget.append(data[i][2:])
budget
# budget1 = list(map(int, budget))
```

```python
# set domestic_gross to the fifth item in the list.
domestic_gross=[]
for i in list(range(4,6000,6)):
    domestic_gross.append(data[i][2:])
domestic_gross
```

```python
# set international_gross to the sixth item in the list.
international_gross=[]
for i in list(range(5,6000,6)):
    international_gross.append(data[i][2:])
international_gross
```

```python
# combine all data above to create a dataframe and drop index after that.
dm=pd.DataFrame(list(zip(index, date, name, budget, domestic_gross, international_gross))).drop(columns=0)
```

```python
# check the shape
dm.shape
```

```python
#sort the table by budget desc.
dm.columns = ['date','name','budget','domestic_gross','international_gross']
dm_sort=dm.sort_values(by='budget', ascending=False)
dm_sort
```

```python
# change str in budget to float
dm['budget'] = dm['budget'].str.replace(',', '').astype(float)
```

```python
# change str in domestic_gross to float
dm['domestic_gross'] = dm['domestic_gross'].str.replace(',', '').astype(float)
```

```python
# change str in international_gross to float
dm['international_gross'] = dm['international_gross'].str.replace(',', '').astype(float)
```

```python
# web scrape again for more data 
def call_individual(web):
    url='https://www.the-numbers.com'+ web
    r=requests.get(url)
    soup=BeautifulSoup(r.content, 'html.parser')
    return soup
```

```python
# find each row of data followed by <b><a><href> to get each url after https://www.the-numbers.com
in_web=[]
for i in range(0,len(list1)):
    for j in list1[i].find_all('tr'):
        if str(type(j.find('b')))!="<class 'NoneType'>":
            in_web.append(j.find('b').find('a').attrs['href'])
in_web
```

```python
# define a function to call each page, will be used later in for loop
def call_eachpage(path):
    url='https://www.the-numbers.com' + path
    r=requests.get(url)
    soup=BeautifulSoup(r.content, 'html.parser')
    return soup
```

```python
# add https://www.the-numbers.com back to make each url the full url and then find all with <tr>
in_web[0]
url='https://www.the-numbers.com'+ in_web[0]
r=requests.get(url)
soup=BeautifulSoup(r.content, 'html.parser')
soup.findAll('tr')
```

```python
# get each movies' running_time from each url and put them in a new list 'genre'
running_time=[]

for web in in_web[:1]:
#     print(in_web)
    soup = call_eachpage(web)
    for i in range(len(soup.findAll('td')[:(len(in_web))])):
#         print(i)
        if soup.findAll('td')[i].text == 'Running Time:':
            running_time.append(call_eachpage(web).findAll('td')[i+1].text[0:-8])
    
    
col_listy=running_time
col_listy
```

```python
# get each movies' genre from each url and put them in a new list 'genre'
genre=[]

for web in in_web[:(len(in_web))]:
#     print(in_web)
    soup = call_eachpage(web)
    for i in range(len(soup.findAll('td')[:(len(in_web))])):
#         print(i)
        if soup.findAll('td')[i].text == 'Genre:':
            genre.append(call_eachpage(web).findAll('td')[i+1].text)
    
    
col_listy2=genre
col_listy2
```

```python
# get each movies' production_company from each url and put them in a new list 'genre'
production_company=[]

for web in in_web[:2]:
#     print(in_web)
    soup = call_eachpage(web)
    for i in range(len(soup.findAll('td')[:(len(in_web))])):
#         print(i)
        if soup.findAll('td')[i].text == 'Production Companies:':
            production_company.append(call_eachpage(web).findAll('td')[i+1].text)
    
    
col_listy3=production_company
col_listy3
```

```python
# get each movies' keywords from each url and put them in a new list 'genre'
keywords=[]

for web in in_web[:2]:
#     print(in_web)
    soup = call_eachpage(web)
    for i in range(len(soup.findAll('td')[:(len(in_web))])):
#         print(i)
        if soup.findAll('td')[i].text == 'Keywords:':
            keywords.append(call_eachpage(web).findAll('td')[i+1].text)
    
    
col_listy4=keywords
col_listy4
```

```python
# change str in running_time to float
dm_sort['running_time'] = dm_sort['running_time'].str.replace(',', '').astype(float)
```

```python
# plot running_time corresponding with international_gross by 4 levels(seperated by quantiles)
ir=dm_sort[['international_gross','running_time']].sort_values(by='running_time', ascending=True)
list10=[ir['international_gross'].iloc[:250].mean(),ir['international_gross'].iloc[250:500].mean(),ir['international_gross'].iloc[500:750].mean(),ir['international_gross'].iloc[750:1000].mean()]
y=list10
x=['level1','level2','level3','level4']
plt.figure(figsize=(10,5))
sns.barplot(x,y)
sns.lineplot(x,y,linewidth=5,color='gold')
plt.title('International Gross vs Running Time')
plt.xlabel('Running Time')
plt.ylabel('International Gross')

number=list(range(0,4))
label=y
for i in range(len(number)):
    plt.text(x=number[i]-0.3, y=label[i]+2, s=round(label[i],2), size=10)


plt.show()
```

```python
# plot budget corresponding with international_gross by 4 levels(seperated by quantiles)
ir2=dm_sort[['international_gross','budget']].sort_values(by='budget', ascending=True)
list11=[ir2['international_gross'].iloc[:250].mean(),ir2['international_gross'].iloc[250:500].mean(),ir2['international_gross'].iloc[500:750].mean(),ir2['international_gross'].iloc[750:1000].mean()]
y=list11
x=['level1','level2','level3','level4']
plt.figure(figsize=(10,5))
sns.barplot(x,y)
sns.lineplot(x,y,linewidth=5,color='#c5c56a')
plt.title('International Gross vs Budget')
plt.xlabel('Budget')
plt.ylabel('International Gross')

number=list(range(0,4))
label=y
for i in range(len(number)):
    plt.text(x=number[i]-0.3, y=label[i]+2, s=round(label[i],2), size=10)


plt.show()
```
